```python
import pandas as pd

df = pd.read_csv("https://drive.google.com/uc?export=download&id=1GW36jlZBoQlQSYOYP4dQc-FCGqUHKPPv")
print(df.head())
```

```python
# Q2. Determine the number of rows and columns in the dataset.

print(df.shape)
```

```python
# Q3. Display all column names.

print(df.columns)
```

```python
# Q4. Identify numerical and categorical features.

print("Numerical Columns:")
print(df.select_dtypes(include=['int64', 'float64']).columns)

print("\nCategorical Columns:")
print(df.select_dtypes(include=['object']).columns)
```

```python
# Q5. Check whether the dataset contains missing values.

print(df.isnull().sum())
```

```python
# Q6. Calculate the average age of users.

print(df['Age'].mean())
```

```python
# Q7. Determine the average watch hours per week.

print(df['WatchHoursPerWeek'].mean())
```

```python
# Q8. Find the average monthly spending of users.

print(df['MonthlySpend'].mean())
```

```python
# Q9. Count the number of users in each subscription category.

print(df['SubscriptionType'].value_counts())
```

```python
# Q10. Determine the percentage of users who renewed their subscriptions.

renewal_percentage = (df['SubscriptionRenewed'] == 'Yes').mean() * 100
print(renewal_percentage)
```

```python
# Q11. Convert categorical features into numerical form.

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_cols = [
    'Gender',
    'SubscriptionType',
    'FavoriteGenre',
    'SubscriptionRenewed'
]

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

print(df.head())
```

```python
# Q12. Define the feature set (X) and target variable (y)
# for subscription renewal prediction.

X = df.drop('SubscriptionRenewed', axis=1)
y = df['SubscriptionRenewed']

print(X.head())
print(y.head())
```

```python
# Q13. Split the dataset into training and testing sets.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
# Q14. Train a Decision Tree model to predict whether a user
# will renew their subscription.

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
```

```python
# Q15. Evaluate the model using accuracy.

from sklearn.metrics import accuracy_score

y_pred_dt = dt.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_dt)

print("Decision Tree Accuracy:", accuracy)
```

```python
# Q16. Generate and interpret the confusion matrix.

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_dt)
print(cm)
```

```python
# Q17. Train a K-Nearest Neighbors (KNN) classifier with K = 5.

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Q18. Compare the accuracy of KNN with the Decision Tree model.

y_pred_knn = knn.predict(X_test)

knn_accuracy = accuracy_score(y_test, y_pred_knn)

print("Decision Tree Accuracy:", accuracy)
print("KNN Accuracy:", knn_accuracy)
```

```python
# Q19. Train a Linear Regression model to predict monthly spending.

from sklearn.linear_model import LinearRegression

X_reg = df.drop('MonthlySpend', axis=1)
y_reg = df['MonthlySpend']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)

lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)
```

```python
# Q20. Predict the monthly spending for a new user and interpret the result.

new_user = [[
    1001, 25, 1, 1, 15, 2, 3, 20, 1
]]

predicted_spend = lr.predict(new_user)

print("Predicted Monthly Spending:", predicted_spend[0])
```
